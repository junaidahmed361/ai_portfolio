### Image Preprocessing

Load Images

```python
# Importing Image class from PIL module
from PIL import Image
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# run a loop to plot the images
def plot_images(images, num_images):
    for i in range(num_images):
        plt.subplot(2, 2, i+1)
        plt.imshow(images[i])
        plt.title('BikeShot' + str(i+1))
        plt.axis('off')
        plt.show()
```

```python
image_path = 'data/'
```

```python
# find number of images in the folder
num_images = len([name for name in os.listdir(image_path) if os.path.isfile(os.path.join(image_path, name))])

# create a list of images
images = []

# loop through the images and append them to the list
for i in range(num_images):
    # rename files to BikeShot1.jpg, BikeShot2.jpg, etc. if file ends with png
    if os.listdir(image_path)[i].endswith('.png'):
        os.rename(image_path + os.listdir(image_path)[i], image_path + 'BikeShot' + str(i+1) + '.jpg')
    images.append(Image.open(image_path + 'BikeShot' + str(i+1) + '.jpg'))
```

```python
images
```

Image Resizing

```python
# resize the images to 416x416
plot_images(images, num_images)
```

Image Normalization and Centering

```python
# Use mean =[0.485, 0.456, 0.406] and std = [0.229, 0.224, 0.225] to "center" your data

# use local centering with the mean and std of the images
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

# loop through the images and normalize them
for i in range(num_images):
    images[i] = images[i].convert('RGB')
    images[i] = np.array(images[i])
    images[i] = images[i] / 255
    images[i] = (images[i] - mean) / std
```

```python
plot_images(images, num_images)
```
